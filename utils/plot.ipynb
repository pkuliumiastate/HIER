```python
import matplotlib.pyplot as plt

# Data for plotting
m_values = [5, 10, 20, 33, 50]  # m values as percentages
# Updated MNIST data for alpha = 1
mnist_values = {
    "FG": [81.86, 81.77, 74.58, 77.24, 75.66],
    "CONTRA": [84.21, 83.59, 82.18, 80.16, 81.64],
    "Ours": [91.99,91.76,92.2,92.32,91.79]  # Updated value for α = 1
}
# Updated CIFAR-10 data for alpha = 1
cifar10_values = {
    "FG": [73.63, 74.84, 72.21, 70.08, 69.11],
    "CONTRA": [76.88, 77.95, 74.66, 72.97, 73.70],
    "Ours": [82.79,82.84,82.43,82.97,81.2]  # Updated value for α = 1
}

# Creating a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plotting MNIST data
ax1.plot(m_values, mnist_values["FG"], label="FG", marker='o', linestyle='-', linewidth=2, color='blue')
ax1.plot(m_values, mnist_values["CONTRA"], label="CONTRA", marker='s', linestyle='-', linewidth=2, color='red')
ax1.plot(m_values, mnist_values["Ours"], label="Ours", marker='^', linestyle='-', linewidth=2, color='green')
ax1.set_title("MNIST Performance at α = 1", fontsize=14)
ax1.set_xlabel("m (%)", fontsize=12)
ax1.set_ylabel("Performance", fontsize=12)
ax1.set_xticks(m_values)
ax1.legend(fontsize=10)
ax1.grid(True)

# Plotting CIFAR-10 data
ax2.plot(m_values, cifar10_values["FG"], label="FG", marker='o', linestyle='-', linewidth=2, color='blue')
ax2.plot(m_values, cifar10_values["CONTRA"], label="CONTRA", marker='s', linestyle='-', linewidth=2, color='red')
ax2.plot(m_values, cifar10_values["Ours"], label="Ours", marker='^', linestyle='-', linewidth=2, color='green')
ax2.set_title("CIFAR-10 Performance at α = 1", fontsize=14)
ax2.set_xlabel("m (%)", fontsize=12)
ax2.set_ylabel("Performance", fontsize=12)
ax2.set_xticks(m_values)
ax2.legend(fontsize=10)
ax2.grid(True)

# Adjusting layout for better appearance
plt.tight_layout()

# Saving the figure (optional)
plt.savefig("performance_comparison.pdf", format='pdf', dpi=300)

# Displaying the combined plot
plt.show()
```
