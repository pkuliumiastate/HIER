```python
"""Plot examples of SciencePlot styles."""

import numpy as np
import matplotlib.pyplot as plt
import scienceplots
pparam = dict(xlabel='Training Epoch', ylabel='Test Accuracy')

with plt.style.context(['science', 'ieee', 'no-latex']):
    fig, ax = plt.subplots()
    for p in [50, 100, 300, 500, 1000]:
        data = np.loadtxt(f'runs_100_to_1000_clients_iid/{p}.csv',delimiter = ',', skiprows=1)
        x = data[:, 1]
        y = data[:, -1]
        ax.plot(x, y, label=p)
    ax.legend(title='# clients')
    ax.autoscale(tight=False)
    ax.set(**pparam)
    # Note: $\mu$ doesn't work with Times font (used by ieee style)
    ax.set_ylabel('Test Accuracy')  
    fig.savefig('figures/100_1000clients_iid.pdf')
    fig.savefig('figures/100_1000clients_iid.jpg', dpi=300)
```

```python
"""Plot examples of SciencePlot styles."""

import numpy as np
import matplotlib.pyplot as plt
import scienceplots
pparam = dict(xlabel='Round', ylabel='Accuracy')

with plt.style.context(['science', 'ieee', 'no-latex', 'std-colors']):
    fig, ax = plt.subplots()
    for p in [100, 95, 90, 80, 67, 50]:
        data = np.loadtxt(f'runs_100clients/{p}.csv',delimiter = ',', skiprows=1)
        x = data[:, 1]
        y = data[:, -1]
        ax.plot(x, y, label=100 - p)
    ax.legend(title='m(%)')
    ax.autoscale(tight=False)
    ax.set(**pparam)
    # Note: $\mu$ doesn't work with Times font (used by ieee style)
    ax.set_ylabel('Accuracy')  
    fig.savefig('figures/runs_100clients.pdf')
    fig.savefig('figures/runs_100clients.eps', dpi=300)
```

```python
"""Plot examples of SciencePlot styles."""

import numpy as np
import matplotlib.pyplot as plt
import scienceplots
pparam = dict(xlabel='Round', ylabel='Accuracy')

with plt.style.context(['science', 'ieee', 'no-latex', 'std-colors']):
    fig, ax = plt.subplots()
    for p in [0, 10, 50, 100, 200, 500]:
        data = np.loadtxt(f'runs_500reference/{p}.csv',delimiter = ',', skiprows=1)
        x = data[:, 1]
        y = data[:, -1]
        ax.plot(x, y, label=p)
        print('%.4f' % y[-1])
    ax.legend(title='# reference')
    ax.autoscale(tight=False)
    ax.set(**pparam)
    # Note: $\mu$ doesn't work with Times font (used by ieee style)
    ax.set_ylabel('Accuracy')  
    fig.savefig('figures/runs_500reference.pdf')
    fig.savefig('figures/runs_500reference.jpg', dpi=300)
```
